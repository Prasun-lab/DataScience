```python
import random

def random_turn():
    return random.choice(["User", "Computer"])

def display_board(board):
    print(" 1 | 2 | 3 ")
    print("-----------")
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("-----------")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("-----------")
    print(f" {board[6]} | {board[7]} | {board[8]} ")

def check_winner(board):
    patterns = [(0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6)]
    for pattern in patterns:
        if board[pattern[0]] == board[pattern[1]] == board[pattern[2]] != " ":
            return True
    return False

def user_turn(board, user_icon):
    while True:
        position = int(input("Enter the position (1-9): "))
        if board[position - 1] == " ":
            board[position - 1] = user_icon
            break
        else:
            print("That position is already taken. Try again.")

def computer_turn(board, computer_icon):
    while True:
        position = random.randint(1, 9)
        if board[position - 1] == " ":
            board[position - 1] = computer_icon
            break

def main():
    board = [" " for _ in range(9)]
    icons = ['X', 'O']
    user_icon = input("Choose your icon (X or O): ").upper()
    computer_icon = 'O' if user_icon == 'X' else 'X'

    current_turn = random_turn()
    print(f"The first turn will be {current_turn}'s turn.")

    while True:
        display_board(board)

        if current_turn == "User":
            user_turn(board, user_icon)
            if check_winner(board):
                display_board(board)
                print("Congratulations! You win!")
                break
        else:
            computer_turn(board, computer_icon)
            if check_winner(board):
                display_board(board)
                print("Computer wins! Better luck next time.")
                break

        if " " not in board:
            display_board(board)
            print("It's a tie!")
            break

        current_turn = "User" if current_turn == "Computer" else "Computer"

if __name__ == "__main__":
    main()
```
